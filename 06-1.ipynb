```python
# !wget https://bit.ly/fruits_300_data -O fruits_300.npy

import urllib.request
urllib.request.urlretrieve(
    "https://bit.ly/fruits_300_data",
    "fruits_300.npy"
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fruits=np.load('fruits_300.npy')
```

```python
print(fruits.shape)
```

```python
# print(fruits[0][0])
print(fruits[0,0,:])
```

```python
plt.imshow(fruits[0],cmap='gray')
plt.show()
```

```python
plt.imshow(fruits[0],cmap='gray_r')
plt.show()
```

```python
fig,axs=plt.subplots(1,2)
axs[0].imshow(fruits[100],cmap='gray_r')
axs[1].imshow(fruits[200],cmap='gray_r')
plt.show()
```

```python
apple=fruits[0:100].reshape(-1,100*100)
pineapple=fruits[100:200].reshape(-1,100*100)
banana=fruits[200:300].reshape(-1,100*100)
```

```python
print(apple.shape)
```

```python
print(apple.mean(axis=1))
```

```python
plt.hist(np.mean(apple,axis=1),alpha=0.8)
plt.hist(np.mean(pineapple,axis=1),alpha=0.8)
plt.hist(np.mean(banana,axis=1),alpha=0.8)
plt.legend(['apple','pineapple','banana'])
plt.show
```

```python
fig,axs=plt.subplots(1,3,figsize=(20,5))
axs[0].bar(range(10000),np.mean(apple,axis=0))
axs[1].bar(range(10000),np.mean(pineapple,axis=0))
axs[2].bar(range(10000),np.mean(banana,axis=0))
# fig1,axs1=plt.subplots(1,3,figsize=(20,5))
# axs1[0].bar(range(100),np.mean(apple,axis=1))
# axs1[1].bar(range(100),np.mean(pineapple,axis=1))
# axs1[2].bar(range(100),np.mean(banana,axis=1))
plt.show()
```

```python
apple_mean=np.mean(apple,axis=0).reshape(100,100)
pineapple_mean=np.mean(pineapple,axis=0).reshape(100,100)
banana_mean=np.mean(banana,axis=0).reshape(100,100)
fig,axs=plt.subplots(1,3,figsize=(20,5))
axs[0].imshow(apple_mean,cmap='gray_r')
axs[1].imshow(pineapple_mean,cmap='gray_r')
axs[2].imshow(banana_mean,cmap='gray_r')
plt.show()
```

```python
abs_diff=np.abs(fruits-apple_mean) #(300, 100, 100)
abs_mean=np.mean(abs_diff,axis=(1,2))
print(abs_mean.shape)
```

```python
apple_index=np.argsort(abs_mean)[:100]
fig,axs=plt.subplots(10,10,figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i,j].imshow(fruits[apple_index[i*10+j]],cmap='gray_r')
        axs[i,j].axis('off')
plt.show()
```

```python
abs_diff=np.abs(fruits-pineapple_mean) #(300, 100, 100)
abs_mean=np.mean(abs_diff,axis=(1,2))
print(abs_mean.shape)
pineapple_index=np.argsort(abs_mean)[:100]
fig,axs=plt.subplots(10,10,figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i,j].imshow(fruits[pineapple_index[i*10+j]],cmap='gray_r')
        axs[i,j].axis('off')
plt.show()
```

```python
abs_diff=np.abs(fruits-banana_mean) #(300, 100, 100)
abs_mean=np.mean(abs_diff,axis=(1,2))
print(abs_mean.shape)
banana_index=np.argsort(abs_mean)[:100]
fig,axs=plt.subplots(10,10,figsize=(10,10))
for i in range(10):
    for j in range(10):
        axs[i,j].imshow(fruits[banana_index[i*10+j]],cmap='gray_r')
        axs[i,j].axis('off')
plt.show()
```
